```python
#import different libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
energy= pd.read_csv(r"C:\Users\mauli\Desktop\dataset\weather_daily_dataset\my_energy_data.csv")
```

```python
energy_data = energy.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', 'energy_median','energy_mean','energy_max','energy_count','energy_std', 'energy_min'])
energy_data
```

```python
#To remove unnecessary columns;
final = energy_data.drop(columns = ['temperatureMax', 'windBearing', 'cloudCover', 'windSpeed', 'pressure', 'visibility', 'humidity', 'moonPhase', 'clear-day', 'cloudy','fog','partly-cloudy-day','partly-cloudy-night','wind','rain', 'snow'])
final
```

```python
#Setting the value for X and Y
x = final.drop('energy_sum', axis = 'columns')
y = final['energy_sum']
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'mse', max_depth = 10, min_samples_split = 10,
                                 random_state = 100)
regressor.fit(x_train, y_train)
```

```python
#Prediction of test set
y_pred_regressor=regressor.predict(x_test)
#Predicted values
print("Prediction for test set: {}".format(y_pred_regressor))
```

```python
#Actual value and the predicted value
regressor_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_regressor})
regressor_diff.head()
```

```python
#Model Evaluation
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_regressor)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_regressor)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_regressor))
print('R squared: {:.2f}'.format(regressor.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
