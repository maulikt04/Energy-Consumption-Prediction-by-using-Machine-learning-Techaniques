```python
#import different libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
energy= pd.read_csv(r"C:\Users\mauli\Desktop\dataset\weather_daily_dataset\my_energy_data.csv")
```

```python
energy_data = energy.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', 'energy_median','energy_mean','energy_max','energy_count','energy_std', 'energy_min'])
energy_data
```

```python
#now first of all find out co-relation with help of Seaborn's heatmap() function:
energy_data.corr()   #personal correlation coefficient:
```

```python
plt.figure(figsize = (30,25))
sns.heatmap(energy_data.corr())
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.show()
```

```python
#Now we can choose different columns according to strong or weak relation between our depemdent and independent columns;
```

```python
# 1) dewpoint, 2) apparentTemperatureHigh, 3) apparentTemperatureLow, 4) apparentTemperatureMax, 5) uvindex
# 6) temperatureLow, 7) temperatureMin, 8) temperatureHigh, 9)apparentTemperatureMin
```

```python
#To remove unnecessary columns;
final = energy_data.drop(columns = ['temperatureMax', 'windBearing', 'cloudCover', 'windSpeed', 'pressure', 'visibility', 'humidity', 'moonPhase', 'clear-day', 'cloudy','fog','partly-cloudy-day','partly-cloudy-night','wind','rain', 'snow'])
final
```

```python
#Setting the value for X and Y
x = final.drop('energy_sum', axis = 'columns')
y = final['energy_sum']
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
#Fitting the Multiple Linear Regression model
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```python
#Prediction of test set
y_pred_mlr= mlr.predict(x_test)
#Predicted values
print("Prediction for test set: {}".format(y_pred_mlr))
```

```python
#Actual value and the predicted value
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
mlr_diff.head()
```

```python
#Model Evaluation
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

```python
#Now we need to check our model accuracy is correct or not, second thing to check our model is overfit or underfit.
#Then we can use "Validation"
#Method: K-Fold cross Validation
```

```python
from sklearn.model_selection import KFold
model = LinearRegression()
kfold_validation = KFold(10)
```

```python
from sklearn.model_selection import cross_val_score
result = cross_val_score(model, x, y, cv = kfold_validation)
```

```python
print(result*100)
print(np.mean(result*100))
```

```python
#Training score by cross validation, here I take number of K-Fold is 25 for training:
train_scores = cross_val_score(model, x_train, y_train , cv = 25) 
train_scores*100
```

```python
#Testing score by cross validation, here I take number of K-Fold is 25 for testing:
test_scores = cross_val_score(model, x_test, y_test , cv = 25) 
test_scores*100
```

```python
#Mean of trining score:
print(np.mean(train_scores*100))
```

```python
#Mean of testing score:
print(np.mean(test_scores*100))
```
