```python
#import different libraries:
import numpy as np
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
daily_data = sorted(glob(r'C:\Users\mauli\Desktop\dataset\daily_dataset/block_*.csv'))
daily_data
```

```python
daily_data_merge = pd.concat(pd.read_csv(datafile).assign(sourcefilename = datafile)
                             for datafile in daily_data )
daily_data_merge
```

```python
daywise = daily_data_merge.drop(['sourcefilename'], axis = 1)
daywise.head()
```

```python
#to fill the empty cell in dataset by using fill forword method and finally prepare this dataset with non-null values: 
daywise_data = daywise.fillna(method = 'ffill')
daywise_data
```

```python
#to load weather daily darksky dataset and to extract date and time:
weather_daily_data = pd.read_csv(r"C:\Users\mauli\Desktop\dataset\weather_daily_dataset\weather_daily_darksky.csv")
weather_daily_data = pd.DataFrame({'temperatureMaxTime':pd.date_range('2011-11-11 23:00:00',periods=882)})
weather_daily_data['day'] = pd.to_datetime(weather_daily_data['temperatureMaxTime']).dt.date
weather_daily_data['time'] = pd.to_datetime(weather_daily_data['temperatureMaxTime']).dt.time
weather_daily_data
```

```python
weather_original = pd.read_csv(r"C:\Users\mauli\Desktop\dataset\weather_daily_dataset\weather_daily_darksky.csv")
weather_original.head(4)
```

```python
weather_updated = pd.concat([weather_daily_data, weather_original], axis = 'columns')
weather_updated
```

```python
#to drop non-require columns by drop function:
weather_drop_time = weather_updated.drop(['temperatureMaxTime', 'temperatureMinTime', 'apparentTemperatureMinTime', 'apparentTemperatureHighTime', 'sunsetTime', 'sunriseTime', 'temperatureHighTime', 'uvIndexTime', 'summary', 'temperatureLowTime', 'apparentTemperatureMaxTime', 'apparentTemperatureLowTime'], axis = 'columns')
```

```python
weather_drop_time
```

```python
#according to weather_drop_time.info(), here Datatype of icon and precipType columns are object and string data:
#now we require to use 'hot_encoding' by get_dummies function:
dummies = pd.get_dummies(weather_drop_time.icon)
dummies
```

```python
preciptype = pd.get_dummies(weather_drop_time.precipType)
preciptype
```

```python
weather_add_1 = pd.concat([weather_drop_time, dummies ], axis = 'columns')
weather_add_1
```

```python
weather_add_2 = pd.concat([weather_add_1, preciptype  ], axis = 'columns')
weather_add_2
```

```python
weather_prefinal = weather_add_2.drop(['precipType', 'icon'], axis = 'columns')
weather_prefinal
```

```python
#to fill empty cell by forword method:
weather_prefinal = weather_prefinal.fillna(method = 'ffill')
weather_prefinal
```

```python
#now we will require to change datatype of date columns before merge:
daywise_data['day'] = pd.to_datetime(daywise_data['day'])
daywise_data
```

```python
weather_prefinal['day'] = pd.to_datetime(weather_prefinal['day'])
weather_prefinal
```

```python
merge_dataset = pd.DataFrame.merge(daywise_data,weather_prefinal , left_on='day', right_on='day', how='inner')
merge_dataset
```

```python
#to also extract date, month and year which is more helpful during the trainnig time:
merge_dataset['year'] = merge_dataset['day'].dt.year
merge_dataset['month'] = merge_dataset['day'].dt.month
merge_dataset['date'] = merge_dataset['day'].dt.day
merge_dataset = merge_dataset.sort_values(by=['day'])
```
